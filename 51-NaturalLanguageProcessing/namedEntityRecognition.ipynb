### Named Entity Recognition

```python
sentence = "The Eiffel Tower was built from 1887 to 1889 by French Engineer Gutsave Eiffel, whose company specialzed in building metal frameworks and structures"
```

```python
import nltk
words=nltk.word_tokenize(sentence)
```

```python
tag_elements= nltk.pos_tag(words)
```

```python
nltk.download("maxent_ne_chunker")
nltk.download('maxent_ne_chunker_tab') 
nltk.download("words")
```

```python
nltk.ne_chunk(tag_elements).draw()
```
